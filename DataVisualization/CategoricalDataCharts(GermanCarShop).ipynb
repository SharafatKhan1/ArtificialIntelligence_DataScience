```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
germanCarShop = {"Cars": ["Audi", "BMW", "Mercedes"], "Frequency": [124, 98, 113]}
germanCarShop
```

```
gcsDataFrame = pd.DataFrame(germanCarShop)
gcsDataFrame
```

```
gcsDataFrame.append({"Cars": "Total", "Frequency": gcsDataFrame["Frequency"].sum()}, ignore_index=True)
```

```
gcsDataFrame.plot.bar(x="Cars", y="Frequency")
```

```
plt.pie(gcsDataFrame["Frequency"], labels=gcsDataFrame["Cars"], autopct="%0.1f%%", radius=1.5, explode=[0, 0.08, 0], shadow=True)
plt.show()
```
